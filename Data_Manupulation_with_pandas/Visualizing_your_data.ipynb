# VISUALIZING YOUR DATA

## HISTOGRAMS

```python
import matplotlib.pyplot as plt
import pandas as pd

dogs = pd.read_csv("dogs.csv")
dogs
```

```python
dogs["height_cm"].hist()
plt.show()
```

```python
dogs["height_cm"].hist(bins=20)
plt.show()
```

```python
dogs["weight_kg"].hist(bins=20)
plt.show()
```

## Bar plots

```python
avg_weight_by_breed = dogs.groupby("breed")["weight_kg"].mean()
print(avg_weight_by_breed)
```

```python
avg_weight_by_breed.plot(kind="bar")
plt.show()
```

```python
avg_weight_by_breed.plot(kind="bar",
                         title= "Mean Weight by Dog Breed")
plt.show()
```

```python
avg_weight_by_breed = dogs.groupby("breed")["weight_kg"].mean()
```

```python
avg_height_by_breed = dogs.groupby("breed")["height_cm"].mean()
```

```python
avg_height_by_breed.plot(kind="bar",
                         title="Mean Height in cm by Dog Breed")
plt.show()
```

```python
sully = {
    "date": ["2019-01-31", "2019-02-28", "2019-03-31", "2019-04-30", "2019-05-31"],
    "weight_kg": [36.1, 35.3, 32.0, 32.9, 32.0],
    "sex": ["F", "M", "M", "F", "F"]
}
```

```python
print(sully)
```

```python
sully_df = pd.DataFrame(sully)
```

```python
print(sully_df)
```

```python
sully_df.to_csv("sully.csv")
```

## Rotating axis labels

```python
sully_df.plot(x="date", y="weight_kg", kind="line", rot=45)
plt.show()
```

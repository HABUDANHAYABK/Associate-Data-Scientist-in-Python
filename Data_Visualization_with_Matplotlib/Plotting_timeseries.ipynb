# PLOTTING TIME SERIES DATA

```python
import pandas as pd

temp= pd.read_csv(r"c:/users/danhaya/arewads/arewads-fellowship/Global_temp_by_city.csv")
print(temp.head())
```

```python
nig_temp = temp[temp['Country'] == 'Nigeria']
print(nig_temp.head())
```

```python
kano_temp = nig_temp[nig_temp['City'] =='Kano']
lagos_temp = nig_temp[nig_temp['City']=='Lagos']


print(kano_temp.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np

#let compare the average temperature of Kano and Lagos Nigeria

# plot
fig, ax=plt.subplots(2, 1)
ax[0].plot(kano_temp['dt'], kano_temp['AverageTemperature'], color='b')
ax[0].plot(kano_temp['dt'], kano_temp['AverageTemperatureUncertainty'], color='b', linestyle='--')


ax[1].plot(lagos_temp['dt'], lagos_temp['AverageTemperature'], color='r')
ax[1].plot(lagos_temp['dt'], lagos_temp['AverageTemperatureUncertainty'], color='r', linestyle='--')


# set labels
ax[0].set_ylabel('Temperature')
ax[1].set_ylabel("Temperature")
ax[1].set_xlabel('Date')

plt.show()
```

# STOCK MARKET US

```python
stocks= pd.read_csv(r"c:/users/danhaya/arewads/arewads-fellowship/Stock_market.csv", parse_dates=['Date'], index_col='Date')
natural_gas = stocks[['Natural_Gas_Price', 'Natural_Gas_Vol.']]
crude_oil = stocks[['Crude_oil_Price', 'Crude_oil_Vol.']]
```

```python
import matplotlib.pyplot as plt

fig, ax=plt.subplots()
ax.plot(natural_gas.index, natural_gas['Natural_Gas_Price'], color='blue')
ax.set_xlabel("Time")
ax.set_ylabel("Natural Gas Price")

ax2= ax.twinx()
ax2.plot(crude_oil.index, crude_oil['Crude_oil_Price'], color='red')
ax2.set_ylabel("Price of Crude oil")

plt.show()
```

```python
twenty_3_naturlgs = natural_gas['29-12-2023': '03-01-2023']
twenty_3_crdoil = crude_oil['29-12-2023': '03-01-2023']
print(twenty_3_naturlgs.head())
print(twenty_3_crdoil.head())
```

```python
fig, ax=plt.subplots()
ax.plot(twenty_3_naturlgs.index, twenty_3_naturlgs['Natural_Gas_Price'], color='blue')
ax.set_xlabel("Time")
ax.set_ylabel("Natural Gas Price", color='blue')
ax.tick_params('y', color='blue')

ax2= ax.twinx()
ax2.plot(twenty_3_crdoil.index, twenty_3_crdoil['Crude_oil_Price'], color='red')
ax2.set_ylabel("Crude oil Price", color='red')
ax2.tick_params('y', color='red')

plt.show()
```

# FUNCTION THAT PLOT TIME-SERIES DATA

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', color=color)
```

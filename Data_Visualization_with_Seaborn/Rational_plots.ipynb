# Rational plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#seaborn has built-in dataset called tips 

tips = sns.load_dataset('tips')
print(tips.head())
```

```python
#Basic Scatter plot

sns.scatterplot(x='total_bill',
                y='tip',
                data=tips)

plt.show()
```

```python
#let's add some customization

sns.scatterplot(x='total_bill',
                y='tip',
                data=tips,
                hue='smoker')
plt.show()
```

```python
#black and red color

hue_color ={
    "Yes": "black",
    "No": "red"
}

sns.scatterplot(x='total_bill',
                y='tip',
                data=tips,
                hue='smoker',
                palette= hue_color)

plt.show()

```

# relplot()

```python
#horizontal plot

sns.relplot(x='total_bill',
            y='tip',
            data=tips,
            kind='scatter',
            col='smoker')

plt.show()
```

```python
#Vertical plot
sns.relplot(x='total_bill',
            y='tip',
            data=tips,
            row='smoker')

plt.show()
```

# Rational plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#seaborn has built-in dataset called tips 

tips = sns.load_dataset('tips')
print(tips.head())
```

```python
#Basic Scatter plot

sns.scatterplot(x='total_bill',
                y='tip',
                data=tips)

plt.show()
```

```python
#let's add some customization

sns.scatterplot(x='total_bill',
                y='tip',
                data=tips,
                hue='smoker')
plt.show()
```

```python
#black and red color

hue_color ={
    "Yes": "black",
    "No": "red"
}

sns.scatterplot(x='total_bill',
                y='tip',
                data=tips,
                hue='smoker',
                palette= hue_color)

plt.show()

```

# relplot()

```python
#horizontal plot

sns.relplot(x='total_bill',
            y='tip',
            data=tips,
            kind='scatter',
            col='smoker')

plt.show()
```

```python
#Vertical plot
sns.relplot(x='total_bill',
            y='tip',
            data=tips,
            row='smoker',
            hue='smoker')

plt.show()
```

```python
print(tips['time'].unique())
```

```python
time_color={
    "Dinner": "black",
    'Lunch': 'red'
}

sns.relplot(x='total_bill',
            y='tip',
            data=tips,
            kind='scatter',
            col='smoker',
            row='time',
            hue='time',
            palette=time_color)

plt.show()
```

```python
print(tips['day'].unique())
```

```python
day_color ={
    "Thur": "black",
    "Fri": "green",
    "Sat": "blue",
    "Sun": "red"
}

sns.relplot(x='total_bill',
            y='tip',
            data=tips,
            kind='scatter',
            hue='day',
            palette=day_color,
            col='day',
            col_wrap= 2,
            col_order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.show()
```

```python
print(tips.columns)
```

```python
# lets try to subgroup with size
sns.relplot(x='total_bill',
                y='tip',
                data=tips,
                kind='scatter',
                size='size',
                style='smoker',
                hue='smoker',
                alpha=0.85)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
print(tips.head())
```

```python
g = sns.relplot(x='total_bill',
            y='tip',
            data=tips,
            kind='scatter',
            hue='smoker',
            size='size')

# set title and labels
g.fig.suptitle("Total Bills vs Tips")
g.set(xlabel='Total Bills',
      ylabel='Tips')



plt.show()
```

```python
g = sns.relplot(x="tip",
            y='total_bill',
            data = tips,
            kind='line',
            hue= 'time',
            style='time'
            )
#set title and label
g.fig.suptitle("TOTAL BILLS VS TIPS")
g.set(xlabel='Tips',
      ylabel='Total Bills')
plt.show()
```

```python
africa20 = pop[pop['Continent']=='Africa']['2020 Population'].sum()
euro20 = pop[pop['Continent']== 'Europe']['2020 Population'].sum()
north20= pop[pop['Continent']=='North America']['2020 Population'].sum()
asia20 = pop[pop['Continent']=='Asia']['2020 Population'].sum()p
oceania20 = pop[pop['Continent']== 'Oceania']['2020 Population'].sum()
south20 = pop[pop['Continent']=='South America']['2020 Population'].sum()
```

```python
continent = pd.DataFrame({
    "countries" : ["Africa", "Europe", "North America", "Asia", "Oceania", "South America"],
    "Population": [africa20, euro20, north20, asia20, oceania20, south20]
})
```

```python
print(continent)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

continent_color ={
    "Africa": "red",
    "Europe": "green",
    "North America": "blue",
    "Asia": "black",
    "South America": "orange",
    "Oceania": "purple"
}

#plot 

g = sns.catplot(x='countries',
            y='Population',
            data = continent,
            kind = 'bar',
            hue= 'countries',
            hue_order= continent_color)

#set label
g.fig.suptitle("Continent with their population")
g.set(xlabel='Countinent',
      ylabel= '2020 Population')

plt.xticks(rotation=45)

plt.show()
```

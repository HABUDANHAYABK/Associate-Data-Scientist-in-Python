# PLOT CUSTOMIZATION

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
print(tips.head())
```

```python
g = sns.relplot(x='total_bill',
            y='tip',
            data=tips,
            kind='scatter',
            hue='smoker',
            size='size')

# set title and labels
g.fig.suptitle("Total Bills vs Tips")
g.set(xlabel='Total Bills',
      ylabel='Tips')



plt.show()
```

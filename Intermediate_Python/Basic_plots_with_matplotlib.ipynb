# DATA VISUALIZATION 

## MATPLOTLIB

```python
import matplotlib.pylab as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)
plt.show()
```

### Adding Title and axis labels

```python
import matplotlib.pylab as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)
#adding x label
plt.xlabel("Year")
# Adding y label
plt.ylabel("Population")
plt.title("Population Graph")
plt.show()
```

### Adding more data

```python
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

year = [1930, 1940] + year
pop = [1.823, 1.882] + pop
plt.plot(year, pop)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Graph")

plt.show()
```

## Scatter plot

```python
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.scatter(year, pop)
plt.show()
```

## Histogram

```python
import matplotlib.pyplot as plt
values = [0, 0.6, 1.4, 1.6, 2.2, 2.5, 2.6, 3.2, 3.5, 3.9, 4.2, 6]
plt.hist(values, bins=3)
```

### TICKS

```python
import matplotlib.pyplot as plt
values = [0, 0.6, 1.4, 1.6, 2.2, 2.5, 2.6, 3.2, 3.5, 3.9, 4.2, 6]
plt.hist(values, bins=3)
plt.yticks([0, 1, 2, 3, 4, 5, 6], [0, "1k", "2k", "3k", "4k", "5k", "6k"])
plt.show()
```

### Changing orientation

```python
import matplotlib.pyplot as plt
values = [0, 0.6, 1.4, 1.6, 2.2, 2.5, 2.6, 3.2, 3.5, 3.9, 4.2, 6]
plt.hist(values, bins=3, orientation= 'horizontal')
```

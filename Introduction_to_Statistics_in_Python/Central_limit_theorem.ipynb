# CENTRAL LIMIT THEOREM

```python
import pandas as pd 
import numpy as np
import scipy.stats as stats

die = pd.Series([1, 2, 3, 4, 5, 6])
#Roll 5 times
samp_5 = die.sample(5, replace=True)
np.mean(samp_5)
```

```python
#Rolling the dice

sample_means =[]
for i in range(10):
    samp_5 = die.sample(5, replace=True)
    sample_means.append(np.mean(samp_5))
print(sample_means)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=10)
plt.show()
```

```python
#let us simulate 100 rolls
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sample_means=[]
for i in range(100):
    samp_5= die.sample(5, replace=True)
    sample_means.append(np.mean(samp_5))

sns.histplot(sample_means, bins=20, kde=True,  color='blue', edgecolor='black')
plt.show()
```

```python
sample_means = []
for i in range(1000):
    samp_5 = die.sample(5, replace=True)
    sample_means.append(np.mean(samp_5))
sns.histplot(sample_means, bins=20, kde=True, color='blue', edgecolor='black')
plt.show()
```

# Standard deviation and the CLT

```python
sample_std =[]
for i in range(1000):
    samp_5 = die.sample(5, replace=True)
    sample_std.append(np.std(samp_5))
sns.histplot(sample_std, bins=20, kde=True, color='blue', edgecolor='black')
plt.show()
```

# Discrete Distributions

```python
import pandas as pd
import numpy as np

die = pd.DataFrame({
    "number": [1, 2, 3, 4, 5, 6]
    })

die["prob"] = np.repeat(1/6, 6)

print(die)
```

```python
#sampling from discrete distributions

Expected_value = np.mean(die["number"])
print(f"The expected mean is {Expected_value}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

rolls_10 =die.sample(10, replace=True)

rolls_10["number"].hist(bins=np.linspace(1, 7, 7))

#display
plt.show()
```

```python
# A bigger sample

rolls_100 = die.sample(100, replace=True)
rolls_100["number"].hist(bins=np.linspace(1,7,7))

#display

plt.show()
```

```python
#lets simulate 1000 rolls

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

rolls_1000 = die.sample(1000, replace=True)

rolls_1000['number'].hist(bins = np.linspace(1,7,7))

#display
plt.show()
```

```python
# law of large numbers
""" 
    as the size of your sample increases, the sample mean will approach the expected mean
    """
print("The expected mean is "+ str(np.mean(die['number'])))
print("The sample mean of 10 rolls is " + str(np.mean(rolls_10['number'])))
print("The sample mean of 100 rolls is " + str(np.mean(rolls_100['number'])))
print("The sample mean of 1000 rolls is " +str(np.mean(rolls_1000['number'])))
```

# Continuous Distributions

```python
# Uniform distribution in python

""" 
    let us simulate the arrival of bus in a bus station. it arrives every 12 interval.
    """

from scipy.stats import uniform
#P(wait time <= 7)
print(uniform.cdf(7,0, 12))
```

```python
# Greater than probabilities
#P(wait time >7)

print(1 - uniform.cdf(7, 0, 12))
```

```python
#P(4<= wait time <= 7)

p4 = uniform.cdf(4, 0, 12)
p7 = uniform.cdf(7, 0, 12)

print(f" P(4<= wait time <= 7) is {(p7 - p4):.2f}")
```

```python
# Generating Random Numbers according to Uniform Distribution

from scipy.stats import uniform

uniform.rvs(0, 5, size= 10)
```

#  MEASURES OF CENTER

```python
import pandas as pd

pop = pd.read_csv(r"c:/users/danhaya/arewads/world-population-dynamics/data/processed_data/processed_data.csv")
```

```python
pop = pop.loc[:, ["Country/Territory", "Continent", "Capital", "2020 Population", "Area (km²)", "Density (per km²)"]]
print(pop.head())
```

```python
# Average population per continent

avg_cont_pop = pop.groupby("Continent")["2020 Population"].mean()

print("The average population of each continent is :")
print(avg_cont_pop)
```

```python
import matplotlib.pyplot as plt

avg_cont_pop.plot(kind = 'bar')
#set labels

plt.xlabel("Continent")
plt.ylabel("Population in Billions")
plt.title("Population by Continent")
plt.xticks(rotation=45)

#Display
plt.show()
```

```python
#mean and median population of each continent
import numpy as np

continent_mean_median = pop.pivot_table(values="2020 Population", index="Continent", aggfunc=[np.mean, np.median])

print(continent_mean_median)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(pop["2020 Population"], bins=50, color='blue', edgecolor='black')

#set labels

plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title("Distribution of world population")

#display

plt.show()
```

```python
# measure of center:

pop20 = pop.loc[:, ["Country/Territory", "Continent", "Capital", "2020 Population", "Area (km²)", "Density (per km²)"]]

africa = pop20[pop20["Continent"] == 'Africa']

print(f"The average population of africa is {africa['2020 Population'].mean():.2f} people")
```

```python
#Distribution of african population

import matplotlib.pyplot as plt

plt.hist(africa["2020 Population"], bins= 50, color='green', edgecolor = 'black')

#set labels
plt.xlabel("2020 Population")
plt.ylabel("Frequency")
plt.title("Distribution of african population 2020")

# display
plt.show()
```

```python
#Exploring mean and median population

europe20_mean_median = pop20[pop20["Continent"] == "Europe"]["2020 Population"].agg([np.mean, np.median])

print("The following are the mean and median population of Europe")
print(europe20_mean_median)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.boxplot(africa["2020 Population"], vert=True, patch_artist=True)

# set label
plt.ylabel("Population")
plt.title("Box plot of African Population")

plt.grid=True

#display
plt.show()
```

```python
import matplotlib.pyplot as plt

continents = pop20["Continent"].unique()
data = [pop20[pop20["Continent"]== cont]['2020 Population'] for cont in continents]

plt.figure(figsize=(10, 6))
plt.boxplot(data, labels = continents, patch_artist=True)

plt.ylabel("Population")
plt.title("Box Population by continent")

plt.xticks(rotation=45)
plt.yscale("log")

#display
plt.show()
```

# POISSON DISTRIBUTION

```python
# simulation of animal adoption

import numpy as np
from scipy.stats import poisson

#if the average number of animal adoption per week is 8, what is the P(# adoptions in week = 5)

prob_5= poisson.pmf(5, 8)
print(f"The probability of adopting 5 animals is {(prob_5 * 100):.2f} %")
```

```python
#what is P(# adoptions in a week <= 5)

prob_th_5 = poisson.cdf(5, 8)
print(f"The probability of adoptions less than 5 animals is {(prob_th_5 *100):.2f}%")
```

```python
# What's probabilty of more 5 adoptions
prob_gth_5 = 1 - poisson.cdf(5, 8)
print(f"The probability of morethan 5 adoptions is {prob_gth_5 *100:.2f}%")
```

```python
#Sampling from poisson distributions

samp = poisson.rvs(8, size=100)
print(samp)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

samp = poisson.rvs(8, size=1000)
sns.histplot(samp, bins=20, kde=True, color='blue', edgecolor='black')
plt.show()
```

```python
sample_mean =[]
for i in range(1000):
    samp = poisson.rvs(8, size=100)
    sample_mean.append(np.mean(samp))
sns.histplot(sample_mean, bins=20, kde=True, color='blue', edgecolor='black')
plt.show()
```

```python
sample_std =[]
for i in range(1000):
    samp = poisson.rvs(8, size= 1000)
    sample_std.append(np.std(samp))
sns.histplot(sample_std, bins=20, kde=True, color='blue', edgecolor='black')
plt.show()
```

```python
sample_var =[]
for i in range(100):
    samp = poisson.rvs(8, size=1000)
    sample_var.append(np.var(samp))
sns.histplot(sample_var, bins=20, kde=True, color='blue', edgecolor='black')
plt.show()
```

# BASIC PROBLEMS

A call center receives an average of 10 calls per hour. What is the probability that it receives exactly 7 calls in a given hour.

```python
#Parameters
#lambda = 10
#P(calls=7)

print(f"The probabilty of receiving exactly 7 calls is {poisson.pmf(7, 10) *100:.2f}%")
```

A websites gets an average of 5 customers support requests per minutes. What is the probabilty that in the next minute, it receives at most 3 requests?

```python
print(f"The probabilty of receiving at most 3 calls is {poisson.cdf(3, 5)*100:.2f}%")
```

A factory produces defective items at an average rate of 2 per day. What is the probability that exactly 4 defective items are produced in a day?

```python
print(f"The probabilty that exactly 4 defectives items are produced is {poisson.pmf(4, 2) *100:.2f}%")
```
